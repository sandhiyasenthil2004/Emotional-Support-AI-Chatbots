```python
from transformers import AutoTokenizer
model_ckpt = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
emotion.reset_format()
```

```python
def tokenize(batch):
  temp =tokenizer(batch['text'], padding=True, truncation=True)
  return temp

print(tokenize(emotion["train"][:5]))
```

```python
emotions_encoded = emotion.map(tokenize, batched=True, batch_size=None)
```

```python
from transformers import AutoModel
import torch

model = AutoModel.from_pretrained(model_ckpt)
```

```python
from transformers import AutoModelForSequenceClassification

num_labels = len(classes)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AutoModelForSequenceClassification.from_pretrained(model_ckpt, num_labels = num_labels).to(device)
```

```python
from transformers import TrainingArguments
batch_size = 64
model_name = "distilbert-finetuned-emotion"

training_args = TrainingArguments(output_dir = model_name,
                                 num_train_epochs=2,
                                 learning_rate = 2e-5,
                                 per_device_train_batch_size= batch_size,
                                 per_device_eval_batch_size = batch_size,
                                  weight_decay=0.01,
                                  evaluation_strategy = 'epoch',
                                  disable_tqdm=False)
```

```python
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(pred):
  labels = pred.label_ids
  preds = pred.predictions.argmax(-1)
  f1 = f1_score(labels, preds, average='weighted')
  acc = accuracy_score(labels, preds)
  return {"accuracy": acc, "f1": f1}
```

```python
from transformers import Trainer

trainer = Trainer(model=model, args=training_args,
                  compute_metrics=compute_metrics,
                  train_dataset=emotions_encoded['train'],
                  eval_dataset=emotions_encoded['validation'],
                  tokenizer=tokenizer)
```

```python
trainer.train()
```

```python
preds_outputs = trainer.predict(emotions_encoded['test'])
preds_outputs.metrics
```

```python
import numpy as np
y_preds = np.argmax(preds_outputs.predictions, axis=1)
y_true = emotions_encoded['test'][:]['label']
```

```python
from sklearn.metrics import classification_report
print(classes)
print(classification_report(y_true, y_preds))
```

```python
emotion_responses = {
    "anger": "I understand that you're feeling angry. Try taking deep breaths, listening to calming music, or stepping away from the situation for a while.",
    "fear": "It's okay to feel scared. Try grounding techniques like focusing on your breathing, talking to a loved one, or writing down your thoughts.",
    "joy": "That's great to hear! Enjoy the moment and spread your happiness to others around you.",
    "love": "Love is a beautiful emotion. Keep nurturing your relationships and expressing your feelings openly.",
    "sadness": "I'm sorry you're feeling this way. Consider talking to a close friend, journaling, or engaging in an activity that makes you feel better.",
    "surprise": "Surprises can be exciting! Embrace the unexpected and see where it takes you."
}
```

```python
def get_emotion_response(text):
    input_encoded = tokenizer(text, return_tensors='pt').to(device)

    with torch.no_grad():
        outputs = model(**input_encoded)

    logits = outputs.logits
    pred = torch.argmax(logits, dim=1).item()
    predicted_emotion = classes[pred]

    response = emotion_responses.get(predicted_emotion, "I'm here for you. Let me know how I can help.")

    return predicted_emotion, response

text = "I am so frustrated with everything today!"
predicted_emotion, response = get_emotion_response(text)
print(f"Predicted Emotion: {predicted_emotion}")
print(f"Response: {response}")
```

```python
import matplotlib.pyplot as plt

log_history = trainer.state.log_history

acc_epochs = []
acc_values = []
f1_epochs = []
f1_values = []

for log in log_history:
    if "epoch" in log:
        epoch = log["epoch"]
        if "eval_accuracy" in log:
            acc_epochs.append(epoch)
            acc_values.append(log["eval_accuracy"])
        if "eval_f1" in log:
            f1_epochs.append(epoch)
            f1_values.append(log["eval_f1"])

if not acc_values or not f1_values:
    for log in log_history:
        if "epoch" in log:
            epoch = log["epoch"]
            if "accuracy" in log:
                acc_epochs.append(epoch)
                acc_values.append(log["accuracy"])
            if "f1" in log:
                f1_epochs.append(epoch)
                f1_values.append(log["f1"])

plt.figure(figsize=(10, 5))
plt.plot(acc_epochs, acc_values, marker='o', label='Evaluation Accuracy')
plt.plot(f1_epochs, f1_values, marker='o', label='Evaluation F1 Score', color='green')
plt.xlabel("Epoch")
plt.ylabel("Metric Value")
plt.title("Evaluation Accuracy and F1 Score per Epoch")
plt.legend()
plt.show()
```
